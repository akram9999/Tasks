```python
import numpy as np
import pandas as pd
df= pd.read_csv("googleplaystore.csv",low_memory=False)
df.head()
```

```python
df['App'] = df['App'].str.lower()
df['Genres'] = df['Genres'].str.lower()
df['Category'] = df['Category'].str.lower()
df['Content Rating'] = df['Content Rating'].str.lower()
df
```

```python
df.info()
```

```python
df["Installs"] = df["Installs"].str.replace("+" ,"").str.replace(",", "") 
df
```

```python

df["Size"]= df["Size"].replace("M","")
df["Reviews"]=pd.to_numeric(df["Reviews"],errors= "coerce")
df["Size"]=pd.to_numeric(df["Size"],errors= "coerce")
df["Installs"]=pd.to_numeric(df["Installs"],errors= "coerce").astype("Int64")
df["Price"]=pd.to_numeric(df["Price"],errors= "coerce")
df




```
